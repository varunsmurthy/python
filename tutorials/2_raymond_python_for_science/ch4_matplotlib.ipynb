# Chapter 4: Plotting with Matplotlib

- pyplot is a layer built upon matplotlib.

## Line plots (of 1D arrays)

```python
from numpy import *
from matplotlib import pyplot

x = arange(0,4*pi,0.1)
y = sin(x)

ll = pyplot.plot(x,y)
pyplot.show()
```

```python
#Adding some annotations to the plot

pyplot.plot(x,y)
pyplot.xlabel("x axis (radians)")
pyplot.ylabel("y axis")
pyplot.title("Sine function")
pyplot.axis([0,4*pi,-2.1,2.1])
pyplot.grid(True)
pyplot.text(4,1.5,"two full cycles")
pyplot.annotate("zero crossing",xy=(2*pi,0),xytext=(2*pi,-1),arrowprops=dict(arrowstyle='->'))

pyplot.show()
```

```python
# plotting multiple curves

z = cos(x)

pyplot.plot(x,y,"r-",linewidth = 1,label = "sine")
pyplot.plot(x,z,"b.",linewidth = 2,label = "cosine")
#pyplot.legend(loc='upper left')
pyplot.legend()
pyplot.xlabel("x axis (radians)")
pyplot.ylabel("y axis")
pyplot.title("Sine and cosine functions")
pyplot.axis([0,4*pi,-3.1,3.1])
pyplot.grid(True)

pyplot.show()
```

```python
#creating subplots

x = arange(0,4*pi,0.1)

a = sin(x)
b = cos(x)
c = tan(x)
d = x

pyplot.figure(figsize=(10,8))
pyplot.subplots_adjust(hspace=0.4,wspace=0.4)

pyplot.subplot(2,2,1)
pyplot.plot(x,a,"r-",linewidth = 1,label = "sine")
pyplot.title("sine function")
pyplot.xlabel("radians")
pyplot.ylabel("y")
pyplot.axis([0,4.1*pi,-2,2])
pyplot.grid(True)
pyplot.legend()

pyplot.subplot(2,2,2)
pyplot.plot(x,b,"k-",linewidth = 2,label = "cosine")
pyplot.title("cosine function")
pyplot.xlabel("radians")
pyplot.ylabel("y")
pyplot.axis([0,4.1*pi,-2,2])
pyplot.grid(True)
pyplot.legend()

pyplot.subplot(2,2,3)
pyplot.plot(x,c,"b-",linewidth = 3,label = "tangent")
pyplot.title("tangent function")
pyplot.xlabel("radians")
pyplot.ylabel("y")
pyplot.axis([0,4.1*pi,-2,2])
pyplot.grid(True)
pyplot.legend()

pyplot.subplot(2,2,4)
pyplot.plot(x,d,"g-",linewidth = 1,label = "linear")
pyplot.title("linear function")
pyplot.xlabel("radians")
pyplot.ylabel("y")
pyplot.axis([0,4.1*pi,0,4.1*pi])
pyplot.grid(True)
pyplot.legend()

pyplot.show()
```

## Plots of 2D arrays

### Contour plots

```python
from numpy import *
from matplotlib import pyplot

x = arange(0,10,0.1)
y = arange(0,9,0.1)

(X,Y) = meshgrid(x,y)
a = sin(X) + cos(Y)

pyplot.contour(x,y,a)
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

```python
# trying to make more sense of the figure using clabel

cont = pyplot.contour(x,y,a)
pyplot.clabel(cont)
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

```python
# trying to show positive and negative values differently without labels

pyplot.contour(x,y,a,arange(-1.5,1.51,0.5),colors = "k")
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

This would be what we'd use if a contour plot had to be overlaid on another color filled contour plot.

```python
#moving on to color-filled contour plots

b = sin(Y) + cos(X)

cont = pyplot.contourf(x,y,b,arange(-2.5,2.5,0.1))
pyplot.colorbar(cont)
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

```python
# finally, overlaying line and colored contour plots

pyplot.jet()
cont = pyplot.contourf(x,y,b,arange(-2.5,2.51,0.1))
pyplot.colorbar(cont)
pyplot.xlabel("x")
pyplot.ylabel("y")

pyplot.contour(x,y,a,arange(-2.5,2.51,0.5), colors = "k", linewidths = 2)

pyplot.show()
```
